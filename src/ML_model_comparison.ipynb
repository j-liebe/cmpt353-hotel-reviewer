```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Assuming the "ML_model_accuracy.csv" file is in the same directory as the notebook.
file_path = "../predicted_data/ML_model_accuracy.csv"
df = pd.read_csv(file_path)
```

```python
# Set a nicer Seaborn style.
sns.set_style("whitegrid")

# Adjust the figure size.
plt.figure(figsize=(9, 6))

# Generate the grouped bar plot using Seaborn.
ax = sns.barplot(data=df, x="Model", y="Accuracy", hue="Prediction Data", palette="pastel", errorbar='sd', capsize=0.1)

# Add labels for each bar showing the accuracy value.
for p in ax.patches:
    ax.annotate(f"{p.get_height():.3f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', color='black', fontsize=9, xytext=(0, 5),
                textcoords='offset points')

# Add labels and title to the plot.
plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.title("Model Accuracy - Training vs. Validation", fontsize=14)

# Position the legend outside the graph.
plt.legend(title="Dataset", fontsize=10, loc='upper left', bbox_to_anchor=(1.02, 1))

# Specify the file path to save the plot.
file_path_output = '../report_related/model_accuracy.png'

plt.ylim(0.8, 1.0)
plt.tight_layout()
plt.savefig(file_path_output, dpi=300)
plt.show()
```
